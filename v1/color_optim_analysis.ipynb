# imports

```python
import os
import pickle
import matplotlib.pyplot as plt
import numpy as np

NCv = 21

class Model:
    def __init__(self, ndn_model, LLs, trial):
        self.ndn_model = ndn_model
        self.LLs = LLs
        self.trial = trial
```

# load models

```python
matt_drifts = []
matt_glms = []
matt_gqms = []
dan_drifts = []
dan_glms = []
dan_gqms = []
for cc in range(NCv):
    matt_drift_filename = 'models/optim/matt_drift_cc'+str(cc)+'.pkl'

    # continue if the file already exists
    if os.path.isfile(matt_drift_filename):
        # load the model and continue
        print('loading model', cc)
        with open(matt_drift_filename, 'rb') as f:
            matt_drifts.append(pickle.load(f))

    matt_glm_filename = 'models/optim/matt_glm_cc'+str(cc)+'.pkl'

    # continue if the file already exists
    if os.path.isfile(matt_glm_filename):
        # load the model and continue
        print('loading model', cc)
        with open(matt_glm_filename, 'rb') as f:
            matt_glms.append(pickle.load(f))

    matt_gqm_filename = 'models/optim/matt_gqm_cc'+str(cc)+'.pkl'

    # continue if the file already exists
    if os.path.isfile(matt_gqm_filename):
        # load the model and continue
        print('loading model', cc)
        with open(matt_gqm_filename, 'rb') as f:
            matt_gqms.append(pickle.load(f))
    
    
    dan_drift_filename = 'models/optim/dan_drift_cc'+str(cc)+'.pkl'
    
    # continue if the file already exists
    if os.path.isfile(dan_drift_filename):
        # load the model and continue
        print('loading model', cc)
        with open(dan_drift_filename, 'rb') as f:
            dan_drifts.append(pickle.load(f))

    dan_glm_filename = 'models/optim/dan_glm_cc'+str(cc)+'.pkl'

    # continue if the file already exists
    if os.path.isfile(dan_glm_filename):
        # load the model and continue
        print('loading model', cc)
        with open(dan_glm_filename, 'rb') as f:
            dan_glms.append(pickle.load(f))

    dan_gqm_filename = 'models/optim/dan_gqm_cc'+str(cc)+'.pkl'

    # continue if the file already exists
    if os.path.isfile(dan_gqm_filename):
        # load the model and continue
        print('loading model', cc)
        with open(dan_gqm_filename, 'rb') as f:
            dan_gqms.append(pickle.load(f))
```

# compare performances

```python
fig = plt.figure(figsize=(12,4))
plt.plot([dan_glm.LLs for dan_glm in dan_glms], label='dan')
plt.plot([matt_glm.LLs for matt_glm in matt_glms], label='matt')
plt.legend()
plt.xticks(range(NCv))
plt.title('GLM LLs')
plt.show()
```

```python
fig = plt.figure(figsize=(12,4))
plt.plot([dan_gqm.LLs for dan_gqm in dan_gqms], label='dan')
plt.plot([matt_gqm.LLs for matt_gqm in matt_gqms], label='matt')
plt.legend()
plt.xticks(range(NCv))
#plt.ylim([-10, 1])
plt.title('GLM LLs')
plt.show()
```
